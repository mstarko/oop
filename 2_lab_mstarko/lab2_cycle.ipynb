```python
mmd = ["червоний", "жовтий", "синій", "зелений", "коричневий", "оранжевий"]
factory = mmd * 100
print(f"На фабриці втготовили {len(factory)} M&Ms {factory}")
from random import shuffle

shuffle(factory)

# розсортовуємо M&Ms по пачкам
pack = []

i = 0 
while i < len(factory):
    pack = factory[i:i+30]
    print(f"Посортували в пачку {len(pack)}: {pack}")
    i += 30
```

```python
while pack:
    print(f"Витягуємо {pack.pop(0)} M&Ms і зїли!")
```

```python
np = int(len(factory) / 30) # скільки пачок вийде

for i in range(np):
    start = i*30
    stop = start + 30
    pack = factory[start:stop]
    print(f"Посортували в пачку {len(pack)}: {pack}")
```

```python
for m in pack:
    print(f"Витягуємо {m} M&Ms і зїли!")
```

```python
from random import randint

for i in range(3):
    dice = randint(1, 6)
    print(f"Кинули кубик і нам випало число: {dice}")
    if dice == 6:
        print("Ура ми виграми!")
        break
    else:
        print("Спробуйте ще.")
```

```python
i = 1
dice = None

while i <= 3 and dice != 6:
    dice = randint(1, 6)
    print(f"Кинули кубик і нам випало число: {dice}")
    i += 1

if dice == 6:
        print("Ура ми виграми!")
else:
    print("Використали всі спроби і програли.")
```

```python
# Виводимо числа від 1 до 5
for i in range(1, 6):
    print(f"Цикл for: {i}")
```

```python
# Виводимо числа від 1 до 5 за допомогою while
i = 1
while i <= 5:
    print(f"Цикл while: {i}")
    i += 1
```

```python
# Використовуємо break для завершення циклу при певній умові
for i in range(1, 10):
    if i == 5:
        print("Зупиняємось на 5")
        break
    print(i)
```

```python
# Пропускаємо виведення числа 3 за допомогою continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```
