```python

class MyContextManager:
    def __enter__(self):
        print("Входимо у контекст")
        return self  # можна повернути об'єкт для використання в блоці

    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type:
            print(f"Виняток: {exc_value}")
        print("Виходимо з контексту")

# Використання
with MyContextManager():
    print("Всередині контексту")
```

```python
from contextlib import contextmanager

@contextmanager
def my_context():
    print("Входимо у контекст")
    yield  # тут можна передати значення у блок `with`
    print("Виходимо з контексту")

# Використання
with my_context():
    print("Всередині контексту")
```

```python
import threading

lock = threading.Lock()

def critical_section():
    with lock:
        # Код, що виконується у захищеному режимі
        print("Виконання критичної секції")
```

```python
import time
from contextlib import contextmanager

@contextmanager
def timer():
    start_time = time.time()
    yield
    end_time = time.time()
    print(f"Час виконання: {end_time - start_time:.4f} секунд")

# Використання
with timer():
    # Тут виконується деяка довготривала операція
    time.sleep(2)  # Симуляція операції, що триває 2 секунди
```
